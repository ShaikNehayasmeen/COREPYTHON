```python
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_columns:
    data[column].fillna(data[column].median(), inplace=True)
```

```python
columns_to_drop = ['Attach', 'Notes']  
data.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
import re
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
data['JobDescription'] = data['JobDescription'].apply(clean_text)
data['JobRequirment'] = data['JobRequirment'].apply(clean_text)
data
```

```python
import pandas as pd

def extract_skills(text):
    
    skills_list = [
        'python', 'java', 'sql', 'javascript', 'c++', 
        'c#', 'html', 'css', 'ruby', 'swift', 
        'go', 'kotlin', 'typescript', 'php', 
        'rust', 'scala', 'objective-c'
    ]
    
   
    text = text.lower()
    

    extracted_skills = [skill for skill in skills_list if skill in text]
    
    return extracted_skills


example_text = "We are looking for a Python developer with experience in Java and SQL."
extracted_skills = extract_skills(example_text)
print("Extracted Skills:", extracted_skills)
```

```python
def normalize_title(title):
    title_mapping = {
        "software eng.": "Software Engineer",
        "software engineer": "Software Engineer",
        "developer": "Software Developer",
        "data scien.": "Data Scientist",
        # Add more mappings as needed
    }
    return title_mapping.get(title.lower(), title)

data['Title'] = data['Title'].apply(normalize_title)
```

```python
def standardize_term(term):
    term_mapping = {
        "ft": "Full-time",
        "full time": "Full-time",
        "pt": "Part-time",
        "part time": "Part-time",
     
    }
    return term_mapping.get(term.lower(), term)

data['Term'] = data['Term'].apply(standardize_term)
```

```python
data['OpeningDate'] = pd.to_datetime(data['OpeningDate'], errors='coerce')
data['Deadline'] = pd.to_datetime(data['Deadline'], errors='coerce')
data['StartDate'] = pd.to_datetime(data['StartDate'], format='%B %d, %Y', errors='coerce')
```

```python
data.columns
```

```python
# Split the 'Location' column into 'City' and 'Country', handling edge cases
location_split = data['Location'].str.split(',', expand=True)
# If there are more than 2 parts, we will only take the first two
location_split = location_split.iloc[:, :2]  # Take only the first two columns
# Assign proper column names
location_split.columns = ['City', 'Country']
# Handle cases where the second column might be NaN (i.e., missing country)
location_split['Country'].fillna('Unknown', inplace=True)
# Assign the split columns back to the original DataFrame
data[['City', 'Country']] = location_split
# Clean up any extra spaces
data['City'] = data['City'].str.strip()
data['Country'] = data['Country'].str.strip()
print(data)
```

```python
def categorize_education(eligibility):
    if 'Bachelor' in eligibility:
        return 'Bachelor’s Degree'
    elif 'Master' in eligibility:
        return 'Master’s Degree'
    elif 'PhD' in eligibility:
        return 'PhD'
    else:
        return 'Other/None'
def categorize_experience(audience):
    if '0-2 years' in audience:
        return 'Entry Level (0-2 years)'
    elif '2-5 years' in audience:
        return 'Mid-Level (2-5 years)'
    elif '5+ years' in audience:
        return 'Senior Level (5+ years)'
    else:
        return 'Other'
data['Education_Level'] = data['Eligibility'].apply(categorize_education)
data['Experience_Range'] = data['Audience'].apply(categorize_experience)
it_jobs = data[data['IT'] == True]
print(it_jobs[['Eligibility', 'Audience', 'Education_Level', 'Experience_Range']])
```
