```python
import pandas as pd
import matplotlib.pyplot as plt
sales_data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [2000, 2500, 2300, 2800, 3000, 3200],
    'Order_Quantity': [50, 65, 55, 70, 75, 80]
})
plt.plot(sales_data['Month'], sales_data['Sales'], marker='o', linestyle='-', color='b')
plt.title('Monthlysales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
plt.hist(sales_data['Order_Quantity'], bins=5, color='Brown', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
customer_data = pd.DataFrame({
    'Age': [25, 30, 22, 45, 33, 29, 40, 50, 38, 28],
    'Spending_Score': [70, 65, 80, 55, 60, 75, 58, 40, 62, 78]
})
plt.hist(customer_data['Age'], bins=5, color='Brown', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
plt.scatter(customer_data['Age'], customer_data['Spending_Score'], color='red')
plt.title('Spending Patterns by Age')
plt.xlabel('Age')
plt.ylabel('Spendingscore')
plt.grid(True)
plt.show()
```

```python
school_data = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math': [85, 78, 92, 88, 76],
    'Science': [90, 85, 89, 94, 80],
    'English': [78, 83, 88, 82, 75]
})
school_data.set_index('Student')[['Math', 'Science', 'English']].plot(kind='bar')
plt.title('Students performance across subjects')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.legend(title='Subjects')
plt.grid(axis='y')
plt.show()
```

```python
school_data[['Math', 'Science', 'English']].plot(kind='box')
plt.title('Marks distribution across subjects')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```
