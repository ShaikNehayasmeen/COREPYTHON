```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y,color="Brown",linewidth=2,marker='*')
plt.title("Line Plot Example")
plt.xlabel("Price of betelnut")
plt.ylabel("Quantity of betelnut")
plt.show()
```

```python
x=['Jan','Feb','Mar','Apr','May','Jun']
y=[10000,20000,15000,12000,18000,35000]
plt.plot(x,y,color="Green",linestyle='dashed',linewidth=2,marker='*')
plt.title("Savings")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['Brown','Violet','Yellow','Green','Blue',],edgecolor='Red')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages=[12,18,14,16,17,18,19,20,22,22,24,12,18,14,35,42,48,42,53,52,54,55,67,62,61,60]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='Brown',edgecolor='black')
plt.title("Age Distribution of customer")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
prog = ['Python', 'Java', 'Ruby', 'C++']
b = [45, 35, 15, 25]
colors = ['Yellow', 'blue', 'green', 'red'] 
plt.pie(b, labels=prog, colors=colors,autopct='%1.1f%%',wedgeprops={'edgecolor': 'Black'})
plt.title("Programming Language Usage")
plt.show()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='Brown',s=100,edgecolor='black')
plt.title("ScatterPlot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.show()
```
