```python
import pandas as pd
import numpy as np

# Original DataFrame
data = pd.DataFrame({
    'Rank': [1, 2, 3, 4, 5],
    'Release Group': ['Mission: Impossible II', 'Gladiator', 'Cast Away', 'What Women Want', 'Dinosaur'],
    '$Worldwide': [546388108.0, 460583960.0, 429632142.0, 374111707.0, 349822765.0],
    '$Domestic': [215409889.0, 187705427.0, 233632142.0, 182811707.0, 137748063.0],
    'Domestic %': [39.4, 40.8, 54.4, 48.9, 39.4],
    '$Foreign': [330978219.0, 272878533.0, 196000000.0, 191300000.0, 212074702.0],
    'Foreign %': [60.6, 59.2, 45.6, 51.1, 60.6],
    'Year': [2000, 2000, 2000, 2000, 2000],
    'Genres': ['Adventure, Action, Thriller', 'Action, Drama, Adventure', 'Adventure, Drama', 'Comedy, Romance', 'Animation, Family, Adventure'],
    'Rating': ['6.126/10', '8.217/10', '7.663/10', '6.45/10', '6.544/10'],
    'Vote_Count': [6741.0, 19032.0, 11403.0, 3944.0, 2530.0],
    'Original_Language': ['en', 'en', 'en', 'en', 'en'],
    'Production_Countries': ['United States of America', 'United Kingdom, United States of America', 'United States of America', 'United Kingdom, United States of America', 'United States of America']
})

# Create 3 rows with NaN values and add them as duplicates
for _ in range(3):
    # Create a new row by selecting random rows from the original DataFrame and introducing NaNs
    new_row = data.sample(1).reset_index(drop=True)  # Random row
    for col in new_row.columns:
        if np.random.rand() > 0.5:  # 50% chance to introduce NaN in this column
            new_row[col] = np.nan
    data = pd.concat([data, new_row], ignore_index=True)

# Check the resulting DataFrame with null values
print(data)
```

```python
# Replace NaN values in the 'Rank' column with the mode (most frequent value)
data['Rank'] = data['Rank'].fillna(data['Rank'].mode()[0])
data
```

```python
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
non_numeric_columns = data.select_dtypes(exclude=[np.number]).columns
data[non_numeric_columns] = data[non_numeric_columns].fillna(data[non_numeric_columns].mode().iloc[0])
print(data)
```
